# DiceDistro: View Distributions of Dice Rolls #

Quickly display the distribution of dice rolls. 

First, you need to import the dicedistro module:

```python
import dicedistro as dice
```

## Define your dice. ##

**dicedistro** contains a class, *Dice*. In order to simulate rolls, you need to define the die or dice that you will be rolling. The only required parameter is list containing the values for each side. Sides can be numbers, text (such as descriptions of a symbol), or a combination of these. If *any* side contains a non-numerical value, all sides will be treated as text when displaying distributions. 

### Numerical Dice. ###
Let's start with a simple example of a typical six-sided die with the numbers 1 through 6. This can be defined as follows: 

```python
d6 = dice.Dice(sides=[1, 2, 3, 4, 5, 6])
```

This defines a simple six-sided die, sometimes referred to as a D6. You can call `print(d6)` to see the name of your die followed by an array containing each of its sides.

```python
print(d6)
```

Notice that the name is 'd6 (numbers)'. This name is generated automatically based on the number of sides and the type (i.e. numbers versus text). You can also see these values directly by calling `d6.kind` and `d6.name`

```python
d6.kind
```

```python
d6.name
```

### Text Dice ###
What happens, then, if you want to generate a die with symbols instead of numbers? In the board game Too Many Bones by [Chip Theory Games](www.chiptheorygames.com), characters have custom attack dice. Four sides have a single hit, one side is a miss (which they call bones), and one side is a double hit. You could define this as follows: 

```python
tmb_attack = dice.Dice(sides=['hit', 'hit', 'hit', 'hit', 'double hit', 'bones'])
print(tmb_attack)
```

Note that the name 'd6 (text)' is not particularly useful. Let's try again with a more descriptive name: 

```python
tmb_attack = dice.Dice(sides=['hit', 'hit', 'hit', 'hit', 'double hit', 'bones'], name='TMB Attack')
print(tmb_attack)
```

Dice only have one fairly limited function, roll(). I will roll the d6 several times to illustrate how it works. 

```python
for i in range(6):
    print(d6.roll())
```

### Visualizing the Distribution ###
In order to get a sense of the die's distribution, you'll need to roll it thousands of times. **DiceDistro** has a function called `roll_dice()` that can perform this for you and display the results. There is only one required parameter, a list of the dice you want to roll (it takes a list with the assumption that you want to see the distribution for rolls from more than one die). In the following example, however, I will manually define all of the parameters the function accepts. Let's see the distribution of rolling three D6, the original roll for determining your characteristics statistics in the pen and paper roleplaying game Dungeons & Dragons: 

```python
dice.roll_dice(dice=[d6, d6, d6], num_rolls=100000, verbose=True, plot=True, title=None)
```

In the first part, you are given a list of the possible values followed by a colon and then the total number of times that value was rolled in the simulation. Following this, in parenthesis, is the percentage of times that value was rolled. 

After this, you are given a graphical representation of this distribution, which is perhaps more immediately informative (although it has slightly fewer details). 

It makes me wonder how my friend's character had at least one 18 in every single game we played...

There seem to be more people who play Settler's of Catan these days. Perhaps it would be good to know the actual distribution of rolling two D6? 

```python
dice.roll_dice(dice=[d6, d6], num_rolls=100000, plot=True)
```

Well then how does it work when we roll dice with text sides instead of numbers? Let's say our character in Too Many Bones is going to roll 3 attack dice. What can be expect to happen? 

```python
dice.roll_dice(dice=[tmb_attack, tmb_attack, tmb_attack], num_rolls=100000, plot=True)
```

### Mixed Dice Types ###
At present, if *any* of the sides on your die are text, then they will all be treated as text. For example, the game King of Tokyo has players roll several dice with the following sides:

```python
kot = dice.Dice(sides=[1, 2, 3, 'lightning bolt', 'heart', 'claw'], name='KoT Dice')
```

Although three of the sides containing numerical values, all sides are treated as text when displaying the distribution:

```python
dice.roll_dice([kot, kot], plot=True)
```

In the future implementation, I would like to make this behavior optional. In King of Tokyo, this would actually be the desired behavior since it matters how many of each value you roll, not their total. In the Too Many Bones example above, though, it might be preferable to change the single and double hits to the numbers 1 and 2 respectively to get a more immediate sense of the actual impact of the rolls.

I'm leaving the next space for experimentation, with another example die description and call to roll the dice. 

```python
my_die = dice.Dice(['side_one', 'side_two', 'side_three'])
dice.roll_dice([my_die, my_die, my_die], verbose=True, plot=False)
```
